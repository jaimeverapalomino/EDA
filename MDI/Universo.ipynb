```python
import pandas as pd 
```

```python
df_op20 = pd.read_csv("OP20200430.TCL.1", sep=',', encoding = "ISO-8859-1", header=None, usecols=[13], converters={13: str},engine='python')
df_D00 = pd.read_csv("D98_20200430_MDI.DAT", encoding = "ISO-8859-1", header=None)
df_op15 = pd.read_csv("OP15200430.TCL.1", sep=',', encoding = "ISO-8859-1", header=None, usecols=[13], converters={13: str},engine='python')
df_D00_MD5=pd.read_csv("D98_20200430_MD5.DAT", encoding = "ISO-8859-1", header=None)
```

```python
df_D00['formato_D00']=df_D00[0].str.slice(0, 14, 1)
df_op20['operacion_op']=df_op20[13]
del df_op20[13]

df_D00_MD5['formato_D00']=df_D00_MD5[0].str.slice(0, 14, 1)
df_op15['operacion_op']=df_op15[13]
del df_op15[13]

df_op20['a']=df_op20['operacion_op'].str.slice(10, 16, 1)
df_op20['b']=df_op20['operacion_op'].str.slice(16, 20, 1)
df_op20['b']=df_op20['b'].str.replace('0','')
df_op20['formato_D00']=''
df_op20.loc[df_op20['b'].astype(int) < 9, 'formato_D00'] = '02J246'+'0'+df_op20['a']+df_op20['b']
df_op20.loc[df_op20['b'].astype(int) > 9, 'formato_D00'] = '02J246'+df_op20['a']+df_op20['b']
df_op20
```
