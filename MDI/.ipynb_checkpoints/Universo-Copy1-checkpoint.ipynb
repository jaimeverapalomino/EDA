```python
import pandas as pd 
```

```python
df_op20 = pd.read_csv("OP20_20200331.CSV", sep=',', encoding = "ISO-8859-1", header=None, usecols=[13], converters={13: str},engine='python')
df_D00 = pd.read_csv("D98_20200331_MDI.DAT", encoding = "ISO-8859-1", header=None)
```

```python
df_D00['operacion']=df_D00[0].str.slice(0, 14, 1)
df_D00['ope_c1']=df_D00[0].str.slice(0, 6, 1)
df_D00['ope_c2']=df_D00[0].str.slice(6, 13, 1)
df_D00['ope_c3']=df_D00[0].str.slice(13, 15, 1)
df_D00['formato_d00']=df_D00[0].str.slice(6, 13, 1)+df_D00[0].str.slice(13, 15, 1)
```

```python
df_op20['operacion']=df_op20[13]
del df_op20[13]
```

```python
df_op20['a']=df_op20['operacion'].str.slice(10, 16, 1)
df_op20['b']=df_op20['operacion'].str.slice(16, 20, 1)
#df_op20['formato_d00']=df_op20['operacion'].str.slice(10, 16, 1)+df_op20['operacion'].str.slice(16, 20, 1)
```

```python
df_op20['b']=  pd.to_numeric(df_op20["b"])
df_op20['formato_d00']=df_op20['a'].astype(str) + df_op20['b'].astype(str)
df_D00["formato_d00"] = pd.to_numeric(df_D00["formato_d00"])
df_op20['formato_d00']=  pd.to_numeric(df_op20["formato_d00"])
```
