```python
import pandas as pd
from decimal import Decimal
```

```python
def decimal_from_value(value):
    return Decimal(value)
```

```python
pd.set_option("display.precision", 4)
pd.options.display.float_format = "{:,.4f}".format


df_D00 = pd.read_csv("BO10_20191230.TCL", decimal=".", header=None, 
                     converters={1: int, 3:str, 6:int, 7:int, 9:str, 11:decimal_from_value, 13:decimal_from_value, 15:decimal_from_value, 16:int, 17:int })

df_D00.rename(columns={11:'ocy_Val'}, inplace=True)
df_D00.rename(columns={4:'prod'}, inplace=True)

df_D00
```

```python

df_D00.groupby('prod')['ocy_Val'].sum()
```
