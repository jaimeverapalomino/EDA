```python
import pandas as pd 
import glob
```

```python
df_D00 = pd.concat([pd.read_csv(f, header=None, encoding = "ISO-8859-1",) for f in glob.glob('*.TXT')], ignore_index = True)
df_D00['sistema']=df_D00[0].str.slice(0, 2, 1)
df_D00['operacion']=df_D00[0].str.slice(2, 14, 1)
df_D00['cctble']=df_D00[0].str.slice(314, 323, 1)
df_D00
```

```python
df_noenBDU = pd.concat([pd.read_csv(f, header=0, sep=';', usecols=range(0, 2), dtype={0: str}, names=["sistema", "operacion"]) for f in glob.glob('*.CSV')], ignore_index = True)
df_noenBDU
```

```python
df_D00.loc[df_D00["operacion"] == "D11810057700"]
```

```python
#esta en D00 y no en BDU
df2=pd.merge(df_D00, df_noenBDU, on=["operacion"],how="inner",indicator=True)
```

```python
df2
```

```python
df_ctactble = df2.iloc[:, 1:4] 
```

```python
df_ctactble.to_csv ('ccteble_x_operacion.csv', index = False, header=True)
```
