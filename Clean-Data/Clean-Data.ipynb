```
# Import data
df = spark.read.csv('dbfs:/FileStore/tables/movie_sample_dataset.txt', encoding='utf-8', header=True)
df.display()
```

```
type(df)
```

```
# Drop useless attributes
df = df.drop('color', 'language')
df.display()
```

```
# Handle text attributes
df = df.fillna(0, subset=['director_name', 'gross', 'budget'])
df.display()
```

```
# Unify countries names
from pyspark.sql.functions import upper, col

df = df.withColumn("country", upper(df.country))
df.display()
```

```
from pyspark.sql.functions import when
df = df.withColumn("country", when(df.country == "USA","UNITED STATES")
                             .when(df.country.isNull() ,"")
                             .otherwise(df.country))

df.display()
```

```
# Bad data entry
df = df.withColumn("director_name", when(df.director_name == "N/A","")
                                   .when(df.director_name.isNull() ,"")
                                   .otherwise(df.director_name))

df = df.withColumn("director_name", when(df.director_name == "Nan","")
                                   .when(df.director_name.isNull() ,"")
                                   .otherwise(df.director_name))

df = df.withColumn("director_name", when(df.director_name == "Null","")
                                   .when(df.director_name.isNull() ,"")
                                   .otherwise(df.director_name))

df = df.withColumn("movie_title", when(df.movie_title == "Â","")
                                 .when(df.movie_title.isNull() ,"")
                                 .otherwise(df.movie_title))

df.display()
```

```
# Handling outliers
df=df.withColumn("gross",df.gross.cast('float'))
df=df.withColumn("duration",df.duration.cast('float'))
df=df.withColumn("budget",df.budget.cast('float'))

df.printSchema()
```

```
df.display()
```

```
df = df.withColumn("duration", when(df.duration <= 10.0,0)
                              .when(df.duration.isNull() ,0)
                              .otherwise(df.duration))

df = df.withColumn("duration", when(df.duration <= 300.0,0)
                              .when(df.duration.isNull() ,0)
                              .otherwise(df.duration))

df = df.withColumn("imdb_score", when(df.imdb_score <= 0.0,0)
                                .when(df.imdb_score.isNull() ,0)
                                .otherwise(df.imdb_score))

df = df.withColumn("title_year", when(df.title_year <= 2010.0,0)
                                .when(df.title_year.isNull() ,0)
                                .otherwise(df.title_year))

df = df.withColumn("gross", when(df.gross == 0, 0)
                           .when(df.gross.isNull() ,0)
                           .otherwise(df.gross))

df = df.withColumn("budget", when(df.budget == 0, 0)
                            .when(df.budget.isNull() ,0)
                            .otherwise(df.budget))

df.display()
```

```
# Normalize data
# spliting actors

### String Split of the column in pyspark
from pyspark.sql.functions import split

split_cols = split(df['actors'], ',')
df = df.withColumn('actor1', split_cols.getItem(0)) \
       .withColumn('actor2', split_cols.getItem(1)) \
       .withColumn('actor3', split_cols.getItem(2))

df = df.drop('actors')

df.display()
```

```
# Adding new feature
# Add a new metric GOB(Gross over Budget)
df = df.withColumn("GOB",when(df.budget !=0,df.gross/df.budget)
                        .otherwise(0))
    
df.display()
```

```
df_top = df.sort(col("GOB").desc()).limit(15)

df_top.display()
```
